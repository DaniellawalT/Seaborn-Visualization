```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

fruit=pd.read_excel("fruit product data-2.xlsx", "Sheet1")
```

```python
fruit
```

```python
new = fruit.groupby(["Product"]).sum().reset_index()
```

```python
sns.barplot(x = new["Product"], y = new["Amount"], data=new)
```

```python
sns.scatterplot(x = new["Product"], y = new["Amount"], data=new, hue="Product")
```

```python
sns.scatterplot(x = new["Product"], y = new["Amount"], data=new, hue="Product", size="Amount")
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x = new["Product"], y = new["Amount"], data=new, hue="Product", size="Amount")
```

```python
sns.lineplot(x = new["Product"], y = new["Amount"], data=new, marker="*").set_title("Line chart for Product")
```

```python
fruit.groupby("Order Date").sum().reset_index()
```

```python
january_sales = fruit.groupby("Order Date").sum().reset_index().loc[0:12]
```

```python
january_sales
```

```python
plt.figure(figsize=(10,5))
plt.grid() #shows the grid
sns.lineplot(x = "Order Date", y = "Amount", data=january_sales, marker="*").set_title("Line chart for january sales")
```
